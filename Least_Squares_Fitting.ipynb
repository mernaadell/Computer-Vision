```
import matplotlib.pyplot as plt
import numpy as np
```

```
def getpixels(x, a, b, c):
    return a * (b * x) + c
```

```
# Making imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

xdata = np.linspace(0, 4, 50)
y = getpixels(xdata, 3, 1.3, 0.5)
np.random.seed(6)
y_noise = 0.3* np.random.normal(size=xdata.size)
ydata = y + y_noise
X = xdata
Y = ydata
from pylab import *
subplot(111, facecolor='black')
plt.plot(X, Y, '_', color='white');
title('image')
savefig('example.png')
show()
```

```
# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)

num = 0
den = 0
for i in range(len(X)):
    num += (X[i] - X_mean)*(Y[i] - Y_mean)
    den += (X[i] - X_mean)**2
m = num / den
c = Y_mean - m*X_mean

print (m, c)
```

```
Y_pred = m*X + c
subplot(111, facecolor='black')
plt.plot(X, Y, '_', color='white');
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='gray') # predicted
title('line detection')
savefig('example2.png')
show()
```
